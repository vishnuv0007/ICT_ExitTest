```python
#Filling missing values
#As noted in EDA, Land and Building are filled with median,
#location, carport, bathroom filled with mode (since discrete numeric)
```

```python
df['Location']=df['Location'].fillna(df['Location'].mode()[0])
df.isnull().sum()
```

```python
df['Bathroom']=df['Bathroom'].fillna(df['Bathroom'].mode()[0])
df.isnull().sum()
```

```python
df['Carport']=df['Carport'].fillna(df['Carport'].mode()[0])
df.isnull().sum()
```

```python
df.shape
```

```python
df['Land']=df['Land'].fillna(df['Land'].median())
df.isnull().sum()
```

```python
df['Building']=df['Building'].fillna(df['Building'].median())
df.isnull().sum()
```

```python
df['Month']=df['Month'].fillna(df['Month'].median())
df.isnull().sum()
```

```python
df['City/Regency']=df['City/Regency'].fillna(df['City/Regency'].mode()[0])
df.isnull().sum()
```

```python
#three columns price, land, building are having very large number of outliers.
#Since we are going to use linear regression models, the outliers are capped
outlier_checking = ['Price','Land','Building']
Q1=df1[outlier_checking].quantile(0.25)
Q3=df1[outlier_checking].quantile(0.75)

IQR=Q3-Q1
outliers_lower=df1[outlier_checking]<(Q1-1.5*IQR)
outliers_upper=df1[outlier_checking]>(Q3+1.5*IQR)

outliers_mild=outliers_upper | outliers_lower
print(df1[outliers_mild])
outliers_lower
```

```python
df1[outlier_checking] = df1[outlier_checking].clip(lower = Q1-1.5*IQR, upper = Q3+1.5*IQR, axis = 1)
df1
```

```python
X=data_encoded.drop('Price',axis=1)
y=data_encoded['Price']
```

```python
#Linear Regression
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(X_train,y_train)#give for learning
y_pred=model.predict(X_test)#give for predictio
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r_value=r2_score(y_test,y_pred)
r_value
```

Since the target column (ie price) is having very large value, the square of error taken will be large and hence, MSE value obtained is a high value.
**R2 value obtained in Linear Regression model is 0.77**

Lasso Regression

```python
from sklearn.linear_model import Lasso
lo=Lasso()
```

```python
model_lo=lo.fit(X_train,y_train)
y_pred_lo=lo.predict(X_test)
y_pred_lo
```

```python
mse=mean_squared_error(y_test,y_pred_lo)
mse
```
